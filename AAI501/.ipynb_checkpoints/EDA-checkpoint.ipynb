```python
###Baseline RFM for the data features
# Calculate recency score based on quantiles, assigning labels from 1 to 5 (1 being lowest recency)
# Creating recency scores based on quantiles using pandas qcut with labels from 1-5
customer_data["recency_score"] = pd.qcut(customer_data['lifetime'], 5, labels=[5, 4, 3, 2, 1])
    
# Calculate frequency score based on rank quantiles, assigning labels from 1 to 5 (1 being lowest frequency)
customer_data["frequency_score"] = pd.qcut(customer_data["total_quantiy"].rank(method="first"), 5, labels=[1, 2, 3, 4, 5])
```

```python
####Combining both of them as per our matrix
# Combine recency and frequency scores to create RFM segment
customer_data['customer_segment'] = customer_data['recency_score'].astype(str) + customer_data['frequency_score'].astype(str)
customer_data.head()
```

```python
## Modelling data
from sklearn.model_selection import train_test_split
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score, fowlkes_mallows_score
from sklearn.model_selection import GridSearchCV

from sklearn.cluster import KMeans, SpectralClustering, DBSCAN

# Evaluation Metrics
from sklearn.metrics import (
    silhouette_score, 
    fowlkes_mallows_score, 
)
```

```python
# Preprocessing
def prepare_data(customer_data):
    # Select features for clustering
    features = ['monetary', 'total_quantity', 'lifetime']
    
    # Prepare the feature matrix
    X = customer_data[features].values
    
    # Standard Scaling
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    return X_scaled
```

```python
# Elbow Method (for different algorithms)
def plot_elbow_methods(X):
    # Elbow Method for K-Means
    kmeans_inertias = []
    k_range = range(1, 11)
    
    for k in k_range:
        kmeans = KMeans(n_clusters=k, random_state=0, n_init=10)
        kmeans.fit(X)
        kmeans_inertias.append(kmeans.inertia_)
    
    # Elbow Method for GMM (using BIC)
    gmm_bic = []
    
    for k in k_range:
        gmm = GaussianMixture(n_components=k, random_state=0)
        gmm.fit(X)
        gmm_bic.append(gmm.bic_)
    
    # Plotting
    plt.figure(figsize=(15, 5))
    
    # K-Means Elbow Plot
    plt.subplot(1, 2, 1)
    plt.plot(k_range, kmeans_inertias, marker='o')
    plt.xlabel('Number of Clusters (K)')
    plt.ylabel('Inertia')
    plt.title('K-Means Elbow Method')
    
    # GMM BIC Plot
    plt.subplot(1, 2, 2)
    plt.plot(k_range, gmm_bic, marker='o')
    plt.xlabel('Number of Clusters')
    plt.ylabel('BIC Score')
    plt.title('GMM Bayesian Information Criterion')
    
    plt.tight_layout()
    plt.show()
```

```python
def evaluate_clustering(X, labels):
    """
    Calculate clustering evaluation metrics.
    
    Parameters:
    -----------
    X : numpy array, scaled data
    labels : array-like, cluster labels
    
    Returns:
    --------
    dict: Evaluation metrics
    """
    try:
        sil_score = silhouette_score(X, labels)
    except:
        sil_score = None
    
    try:
        fm_score = fowlkes_mallows_score(np.zeros(len(labels)), labels)
    except:
        fm_score = None
    
    return {
        'silhouette_score': sil_score,
        'fowlkes_mallows_score': fm_score
    }
```

```python
def kmeans_clustering(X, n_clusters):
    """
    Perform K-means clustering.
    
    Parameters:
    -----------
    X : numpy array, scaled data
    n_clusters : int, number of clusters
    
    Returns:
    --------
    dict: Clustering results and metrics
    """
    kmeans = KMeans(n_clusters=n_clusters, random_state=0)
    labels = kmeans.fit_predict(X)
    
    return {
        'labels': labels,
        'centroids': kmeans.cluster_centers_,
        'inertia': kmeans.inertia_,
        'metrics': evaluate_clustering(X, labels)
    }

def gmm_clustering(X, n_components):
    """
    Perform Gaussian Mixture Model clustering.
    
    Parameters:
    -----------
    X : numpy array, scaled data
    n_components : int, number of components
    
    Returns:
    --------
    dict: Clustering results and metrics
    """
    gmm = GaussianMixture(n_components=n_components, random_state=0)
    labels = gmm.fit_predict(X)
    
    return {
        'labels': labels,
        'means': gmm.means_,
        'covariances': gmm.covariances_,
        'metrics': evaluate_clustering(X, labels)
    }

def spectral_clustering(X, n_clusters):
    """
    Perform Spectral Clustering.
    
    Parameters:
    -----------
    X : numpy array, scaled data
    n_clusters : int, number of clusters
    
    Returns:
    --------
    dict: Clustering results and metrics
    """
    spectral = SpectralClustering(n_clusters=n_clusters, random_state=0)
    labels = spectral.fit_predict(X)
    
    return {
        'labels': labels,
        'metrics': evaluate_clustering(X, labels)
    }

def dbscan_clustering(X, eps, min_samples):
    """
    Perform DBSCAN clustering.
    
    Parameters:
    -----------
    X : numpy array, scaled data
    eps : float, maximum distance between two samples
    min_samples : int, minimum number of samples in a neighborhood
    
    Returns:
    --------
    dict: Clustering results and metrics
    """
    dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    labels = dbscan.fit_predict(X)
    
    return {
        'labels': labels,
        'metrics': evaluate_clustering(X, labels)
    }
```

```python
def find_best_clustering(X):
    """
    Perform grid search across different clustering algorithms.
    
    Parameters:
    -----------
    X : numpy array, scaled data
    
    Returns:
    --------
    dict: Best clustering results for each algorithm
    """
    results = {}
    
    # K-means grid search
    kmeans_results = []
    for n_clusters in range(4, 11):
        kmeans_result = kmeans_clustering(X, n_clusters)
        kmeans_results.append({
            'n_clusters': n_clusters,
            **kmeans_result
        })
    results['kmeans'] = max(kmeans_results, key=lambda x: x['metrics']['silhouette_score'])
    
    # GMM grid search
    gmm_results = []
    for n_components in range(4, 11):
        gmm_result = gmm_clustering(X, n_components)
        gmm_results.append({
            'n_components': n_components,
            **gmm_result
        })
    results['gmm'] = max(gmm_results, key=lambda x: x['metrics']['silhouette_score'])
    
    # Spectral Clustering grid search
    spectral_results = []
    for n_clusters in range(4, 11):
        spectral_result = spectral_clustering(X, n_clusters)
        spectral_results.append({
            'n_clusters': n_clusters,
            **spectral_result
        })
    results['spectral'] = max(spectral_results, key=lambda x: x['metrics']['silhouette_score'])
    
    # DBSCAN grid search
    dbscan_results = []
    for eps in [0.1, 0.3, 0.5, 1.0]:
        for min_samples in [2, 3, 4, 5]:
            try:
                dbscan_result = dbscan_clustering(X, eps, min_samples)
                # Skip if all points are in one cluster
                if len(np.unique(dbscan_result['labels'])) > 1:
                    dbscan_results.append({
                        'eps': eps,
                        'min_samples': min_samples,
                        **dbscan_result
                    })
            except:
                continue
    
    if dbscan_results:
        results['dbscan'] = max(dbscan_results, key=lambda x: x['metrics']['silhouette_score'])
    
    return results
```

```python
def main(df):
    """
    Main function to perform clustering analysis.
    
    Parameters:
    -----------
    df : pandas DataFrame with customer data
    
    Returns:
    --------
    dict: Clustering results for different algorithms
    """
    # Prepare data
    X = prepare_data(df)
    
    # Find best clustering across algorithms
    results = find_best_clustering(X)
    
    return results
```
