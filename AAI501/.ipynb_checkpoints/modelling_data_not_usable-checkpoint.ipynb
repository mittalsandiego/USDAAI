```python
#Droping null values in occupation column
adult_df = adult_df.dropna(subset=['occupation'])
adult_df.isnull().sum()
```

```python
###Workclass with NULL values is also dropped for the entire dataset if we are dropping Occupation categories
adult_df['native_country'] = adult_df['native_country'].fillna(adult_df['native_country'].mode()[0])
```

```python
#Checking for null values
adult_df.isnull().sum()
```

```python
###Checking for categorical features
cat_feat = adult_df.dtypes==object
cat_feat
```

```python
###Encoding the categorical features identified above features.The assumption is all features are non relational
from sklearn.preprocessing import LabelEncoder

cat_list = list(cat_feat[cat_feat == True].index)
numeric_list = list(cat_feat[cat_feat == False].index)
le = LabelEncoder()
adult_df[cat_list]=adult_df[cat_list].apply(lambda x:le.fit_transform(x))
```

```python
adult_df.head()
```

```python
#### Data Preparation
#Slicing dataset into Independent(X) and Target(y) varibles
y = adult_df.pop('income')
X = adult_df
```

```python
#Scaling the dependent variables
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
#Dividing dataset into test and train
from sklearn.model_selection import train_test_split
ran_state = 0
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = ran_state)

print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

```python
### Modelling
###The aim is to try out multiple models to figure out the best baseline model and therbey hyperparameter tuning of the final model.
###This ensures we get the best model and parameters for the the dataset.More reference of free - lunch theorem\
```

https://en.wikipedia.org/wiki/No_free_lunch_theorem \
https://machinelearningmastery.com/no-free-lunch-theorem-for-machine-learning/

```python
classifier_lg = LogisticRegression(random_state=ran_state)
classifier_dt = DecisionTreeClassifier(random_state=ran_state)
classifier_nb = GaussianNB()
classifier_rf = RandomForestClassifier(random_state=ran_state)
classifier_xgb = XGBClassifier(random_state=ran_state)
```

```python
# Instantiate the classfiers and make a list
classifiers = [classifier_lg,
               classifier_dt,
               classifier_nb,
               classifier_rf,
               classifier_xgb]
# Define a result table as a DataFrame
result_table = pd.DataFrame(columns=['classifiers', 'fpr','tpr','Roc Auc','Accuracy','f1 Score','logloss','Confusion Matrix','Precision','Recall'])
```

```python
# Train the models and record the results
for cls in classifiers:
    model = cls.fit(X_train, y_train)
    y_proba = model.predict_proba(X_test)[::,1]
    y_pred = model.predict(X_test)
    # print(cls, '\n','Confusion Matrix','\n',confusion_matrix(y_test,  y_pred))
    # print('\n','Classification Report','\n',classification_report(y_test,  y_pred))
    # print('='*170)
    fpr, tpr, _ = roc_curve(y_test,  y_proba)
    auc = roc_auc_score(y_test, y_proba)
    Accuracy = accuracy_score(y_test,y_pred)
    f1score = f1_score(y_test,y_pred)
    logloss = log_loss(y_test,y_proba)
    cm = confusion_matrix(y_test,  y_pred)
    precision = precision_score(y_test,  y_pred)
    recall = recall_score(y_test,  y_pred)

    output_metrics = pd.DataFrame({
    'classifiers': [cls.__class__.__name__],
    'fpr': [fpr],
    'tpr': [tpr],
    'Roc Auc': [auc],
    'Accuracy': [Accuracy],
    'f1 Score': [f1score],
    'logloss': [logloss],
    'Confusion Matrix': [cm],
    'Precision': [precision],
    'Recall': [recall]
    })
    
    result_table = pd.concat([result_table, output_metrics], ignore_index=True)

# Set name of the classifiers as index labels
result_table.set_index('classifiers', inplace=True)
```

```python
result_table.to_csv('model_output.csv')
```

##### There are 3 major hyperparameter tuning paradigms:
1. GridSearchCV: Exhaustively considers all parameter combinations
2. RandomizedSearchCV: Sample a given number of candidates from a parameter space with a specified distribution
3. Bayesian Optimization: It is a sequential design strategy for global optimization of black-box functions, that does not assume any functional forms.

https://scikit-learn.org/stable/modules/grid_search.html \
https://en.wikipedia.org/wiki/Bayesian_optimization

We will use bayesian optimization with hyperopt library get the ideal hyperparameter

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
import xgboost as xgb 
dtrain = xgb.DMatrix(data=X_train, label=y_train, 
                     enable_categorical=True)

dtest = xgb.DMatrix(data=X_test, label=y_test, 
                    enable_categorical=True)
```

```python
import gc
gc.collect()
```

```python
from hyperopt import STATUS_OK, Trials, fmin, hp, tpe, space_eval 
from hyperopt.early_stop import no_progress_loss

search_space = {'max_depth' : hp.choice('max_depth', range(5, 30, 1)),
  'learning_rate' : hp.quniform('learning_rate', 0.01, 0.5, 0.01),
  'n_estimators' : hp.choice('n_estimators', range(20, 205, 10)),
  'gamma' : hp.quniform('gamma', 0, 0.50, 0.01),
  'min_child_weight' : hp.quniform('min_child_weight', 1, 10, 1),
  'subsample' : hp.quniform('subsample', 0.1, 1, 0.01),
  'colsample_bytree' : hp.quniform('colsample_bytree', 0.1, 1.0, 0.01),
  'scale_pos_weight': hp.uniform('scale_pos_weight', 0.2, 0.8),
  'objective' : 'binary:logistic',
    'eval_metric' : 'auc',
    'seed' : 0
}
```

```python
def xgb_objective(space):
  results = xgb.cv(space, 
                   dtrain=dtrain, #DMatrix (xgboost specific)
                   num_boost_round=500, 
                   nfold=5, 
                   stratified=True,  
                   early_stopping_rounds=20,
                   metrics = ['logloss','auc','aucpr','error'])
  
  best_score = results['test-auc-mean'].max()
  return {'loss':-best_score, 'status': STATUS_OK}
```

```python
trials = Trials()

best_hyperparams = fmin(fn=xgb_objective, 
                        space=search_space,
                        algo=tpe.suggest,
                        max_evals=500,
                        trials=trials, 
                        return_argmin=False, 
                        early_stop_fn=no_progress_loss(10))

best_params = best_hyperparams.copy()

# `eval_metric` is a key that is not a hyperparameter of the classifier
if 'eval_metric' in best_params:
  best_params = {key:best_params[key] for key in best_params if key!='eval_metric'}

best_params
```

```python
best_model = XGBClassifier(**best_params)
```

```python
best_model.fit(X_train, y_train)
y_proba = best_model.predict_proba(X_test)[::,1]
y_pred = best_model.predict(X_test)
```

```python
auc = roc_auc_score(y_test, y_proba)
print(auc)
```

```python
fpr, tpr, _ = roc_curve(y_test,  y_proba)
auc = roc_auc_score(y_test, y_proba)
Accuracy = accuracy_score(y_test,y_pred)
f1score = f1_score(y_test,y_pred)
logloss = log_loss(y_test,y_proba)
cm = confusion_matrix(y_test,  y_pred)
precision = precision_score(y_test,  y_pred)
recall = recall_score(y_test,  y_pred)
```

```python
output_metrics = pd.DataFrame({
'classifiers': [best_model.__class__.__name__],
'fpr': [fpr],
'tpr': [tpr],
'Roc Auc': [auc],
'Accuracy': [Accuracy],
'f1 Score': [f1score],
'logloss': [logloss],
'Confusion Matrix': [cm],
'Precision': [precision],
'Recall': [recall]
})
    
# result_table = pd.concat([result_table, output_metrics], ignore_index=True)

# Set name of the classifiers as index labels
output_metrics.set_index('classifiers', inplace=True)
```
